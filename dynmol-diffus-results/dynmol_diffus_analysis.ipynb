```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv, sys, time, os, re
import scipy.optimize
from math import *
Ï€ = np.pi
```

```python
path = "/run/media/xif/BigData/BrownianMotion/res51"
#pathsave = path
pathsave = "res51"
D = 0.08

df = pd.read_csv(path+"/params.csv", sep=',', header=None)
df = df.set_index(0).T
params = {}
for key in df:
    params[key] = df[key][1]
if 'delta_t_beg_cutoff' not in params:
    params['delta_t_beg_cutoff'] = 0.10
params
```

```python
realizations = []
for file in os.listdir(path):
    itempath = path+"/"+file
    r0 = re.match('^([0-9]+)$', file)
    if r0 is not None:
    
        d = { 'path' : itempath }
        df = pd.read_csv(itempath+"/res.csv", sep=',', header=None)
        df = df.set_index(0).T
        for key in df:
            d[key] = df[key][1]
        if 't_beg_exact' in d:
            df = pd.read_csv(itempath+"/p_traj.csv", sep=',')
            d['t'] = np.array(df['t'])
            d['x'] = np.array(df['px'])
            d['y'] = np.array(df['py'])
            realizations.append(d)
        else:
            print("fixed acq beg ",itempath)
        
len(realizations)
```

```python
T = []
for rea in realizations:
    if rea['T'] - params['target_T'] > 5:
        print(rea['path'])
    T.append( rea['T'] )
T = np.array(T)
T_mean = np.mean(T)
plt.hist(T)
plt.axvline(x = T_mean, color='black', lw=2)
T_window = 1.8*T_mean/350
plt.axvline(x = T_mean+T_window, linestyle='--', color='black')
plt.axvline(x = T_mean-T_window, linestyle='--', color='black')
None
```

On Ã©limine toutes les trajectoires pour lesquelles $T$ dÃ©vie de plus de $T_\text{window}$ ($\simeq 1Â°$) de la moyenne. On vÃ©rifie que Ã§a n'introduit pas de biais de sÃ©lection (la dÃ©viation de $T$ n'est pas une consÃ©quence de la diffusion).

```python
rea_sel = []
T_sel = []
T_std = []
for k in range(len(realizations)):
    if np.abs(T[k]-T_mean) < T_window:
        rea_sel.append(realizations[k])
        T_sel.append(T[k])
        T_std.append(realizations[k]['T_std'])
realizations = rea_sel
T_mean = np.mean(T_sel)
T_err = T_window + np.mean(T_std)
print("T = {:.2f} Â± {:.2f} for {} realizations".format(T_mean,T_err,len(realizations)))
```

Pas mal de trajectoires qui se terminent tÃ´t (la particule a touchÃ© le bord), et c'est embÃªtant si on veut exploiter toutes les donnÃ©es et pas seulement le dÃ©but des trajectoires. Solution pas trop mauvaise : on extrapole/complÃ¨te les trajectoires comme si la particule continuait Ã  diffuser en $\sqrt{t}$ avec le coeff de diffusion aprroximatif mesurÃ©. On vÃ©rifie a posteriori que le comportement est bien globalement diffusif, et de toute faÃ§on Ã§a concerne un nombe minoritaire de trajectoires. On ne peut pas les Ã©liminer car Ã§a introduirait Ã©videmment un biais de sÃ©lection.

```python
for rea in realizations:
    rea['x0'] = None
    rea['y0'] = None

t = np.linspace(0.02, params['t_end']-params['t_beg'], 150)
samples_t_i = [30, 75, 120]
samples_acc_R2 = []
samples_acc_dir = []
R2 = np.zeros(150)
R2std = np.zeros(150)
mass_center_x = np.zeros(150)
mass_center_y = np.zeros(150)
Nincomplet = np.zeros(150)
for i in range(150):
    acc_R2 = []
    acc_dir = []
    record_sample = (i in samples_t_i)
    mass_center_x_acc, mass_center_y_acc = 0,0
    for rea in realizations:
        k = np.searchsorted(rea['t'], rea['t_beg_exact']+t[i], side='right')
        if rea['x0'] is None:
            rea['x0'] = rea['x'][k]
            rea['y0'] = rea['y'][k]
        if k >= len(rea['t']):
            k = -1
        x = rea['x'][k]-rea['x0']
        y = rea['y'][k]-rea['y0']
        mass_center_x_acc += x
        mass_center_y_acc += y
        r2 = x**2 + y**2
        if k == -1:
            # on simule une diffusion normale Ã  partir de la fin de la traj
            r2 += 4*D * (rea['t_beg_exact']+t[i] - rea['t'][-1])
            Nincomplet[i] += 1
        acc_R2.append(r2)
        if record_sample:
            theta = atan2( y, x )
            acc_dir.append(theta)
    if record_sample:
        samples_acc_R2.append(acc_R2)
        samples_acc_dir.append(acc_dir)
    mass_center_x[i] = mass_center_x_acc/len(realizations)
    mass_center_y[i] = mass_center_y_acc/len(realizations)
    R2[i] = np.mean(acc_R2)
    R2std[i] = np.std(acc_R2)
Nincomplet /= len(realizations)
t -= t[0]
```

[LEGACY] Fit d'une droite $\langle r^2\rangle=4D(t-t_0)$

```python
t0, t1 = 0.01, 25/T_mean
i0 = np.searchsorted(t, t0, side='left')
i1 = np.searchsorted(t, t1, side='right')-1
_coeff, _cov = scipy.optimize.curve_fit( (lambda t,D: 4*D*t), t[i0:i1]-t[i0], R2[i0:i1] )
D = _coeff[0]
D_err = sqrt(np.diag(_cov)[0])
print("straight fit on {:.3f}->{:.3f} : D = {:.4f} Â± {:.1f}%fit Â± {:.1f}%stat".format(t[i0],t[i1],D,D_err/D*100,150/sqrt(len(realizations))))
D_err += D * 1.5/sqrt(len(realizations))
```

Fit de la courbe complÃ¨te incluant le rÃ©gime inertiel, sur $\tau$ et $D$ (*avec* dispersion initiale de la vitesse (`diffus_curve_v0disp` de `langevin.ipynb`))

```python
m = params['part_m']
diffus_curve = lambda t,tau,T: 4*T/m*tau * ( t -tau*(1-np.exp(-t/tau)) )

p0 = (0.01, T_mean)
R2_err = 3 * R2std / sqrt(len(realizations))
R2_err[0] = R2_err[1]
_coeff, _cov = scipy.optimize.curve_fit( diffus_curve, t[:i1]-t[0], R2[:i1], p0=p0, sigma=1.5*R2_err[:i1], absolute_sigma=True, bounds=([p0[0]*0.5,p0[1]*0.5],[p0[0]*1.5,p0[1]*1.5]) )
diffus_tau = _coeff[0]
diffus_tau_err = sqrt(np.diag(_cov)[0])
diffus_T = _coeff[1]
diffus_T_err = sqrt(np.diag(_cov)[1])

D = diffus_T/m*diffus_tau
D_err = D * (diffus_tau_err/diffus_tau + diffus_T_err/diffus_T)

plt.figure(figsize=(9,6))
ax1 = plt.gca()
ax1.fill_between(t, R2-R2_err, R2+R2_err, facecolor='orange', alpha=0.5, label=r"$\pm \operatorname{std}(r^2)\cdot 3/\sqrt{n}$")
ax1.plot(t, R2, label=r"$\langle r^2 \rangle_\operatorname{ens}(t)$", lw=2)
ax1.plot(t[:i1], diffus_curve(t-t[0],diffus_tau,diffus_T)[:i1], '--', color='black', label=r"fit $D={:.3f}\pm{:.3f}$".format(D,D_err))
ax1.plot(t[:i1], diffus_curve(t-t[0],diffus_tau-diffus_tau_err,diffus_T-diffus_T_err)[:i1], '--', color='grey', lw=1)
ax1.plot(t[:i1], diffus_curve(t-t[0],diffus_tau+diffus_tau_err,diffus_T+diffus_T_err)[:i1], '--', color='grey', lw=1)
#ax1.plot(t[:i1], 4*t[:i1]*D)
ax1.legend()
ax1.set_xlabel("$t$")
ax1.set_ylabel("$r^2$")
plt.title("Diffusion for {} trajectories, @ $T={:.1f}\pm{:.1f}$, $m_\operatorname{{part}}={}$, $d_\operatorname{{part}}={}$".format(len(realizations), T_mean, T_err, params['part_m'], params['part_d']))
if path is not None:
    plt.savefig(pathsave+"/diffus_fit.pdf", bbox_inches='tight')

print("fit : D = {:.4f} Â± {:.1f}%".format(D,D_err/D*100))
print("fit : tau = {:.4f} Â± {:.1f}%".format(diffus_tau,diffus_tau_err/diffus_tau*100))
gamma = m / diffus_tau
gamma_err = gamma * diffus_tau_err/diffus_tau

pd.DataFrame({
    't': t,
    'R2': R2,
    'R2std': R2std
}).to_csv(pathsave+"/R2_t.csv", index=False, float_format='%.5f')
```
